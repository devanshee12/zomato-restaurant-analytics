```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/Pan-India Zomato Restaurant Review Analytics/zomato.csv")

print("Before Cleaning → Rows, Cols:", df.shape)

# 1. Standardize column names (lowercase, underscores, no spaces)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", "")

# ✅ Columns now look like:
# ['index','restaurant_name','restaurant_type','rate_out_of_5','num_of_ratings','avg_cost_two_people',
# 'online_order','table_booking','cuisines_type','area','local_address']

# 2. Drop duplicates (same restaurant appearing multiple times)
df.drop_duplicates(inplace=True)

# 3. Handle missing values
df = df.dropna(subset=['restaurant_name', 'rate_out_of_5', 'area'])   # drop critical nulls
df['num_of_ratings'] = df['num_of_ratings'].fillna(0).astype(int)
df['avg_cost_two_people'] = df['avg_cost_two_people'].fillna(df['avg_cost_two_people'].median())

# 4. Clean rating column
df['rate_out_of_5'] = df['rate_out_of_5'].astype(str).str.extract(r'(\d+\.?\d*)')   # extract numbers
df['rate_out_of_5'] = pd.to_numeric(df['rate_out_of_5'], errors='coerce')
df = df[df['rate_out_of_5'].between(0,5)]   # only keep valid ratings

# 5. Clean cost column (remove symbols, commas, convert to int)
df['avg_cost_two_people'] = df['avg_cost_two_people'].astype(str).str.replace("[^0-9]", "", regex=True)
df['avg_cost_two_people'] = pd.to_numeric(df['avg_cost_two_people'], errors='coerce')
df['avg_cost_two_people'] = df['avg_cost_two_people'].fillna(df['avg_cost_two_people'].median()).astype(int)

# 6. Standardize categorical values (online_order, table_booking)
df['online_order'] = df['online_order'].str.strip().str.lower().replace({'yes':'Yes','no':'No'})
df['table_booking'] = df['table_booking'].str.strip().str.lower().replace({'yes':'Yes','no':'No'})

# 7. Fix cuisines (take only first cuisine if multiple are listed)
df['cuisines_type'] = df['cuisines_type'].astype(str).str.split(',').str[0].str.strip()

# 8. Clean text columns
df['restaurant_name'] = df['restaurant_name'].str.strip().str.title()
df['area'] = df['area'].str.strip().str.title()

# Final check
print("After Cleaning → Rows, Cols:", df.shape)
print(df.dtypes)

# Save cleaned data for Tableau
df.to_csv("/content/drive/MyDrive/Pan-India Zomato Restaurant Review Analytics/cleaned_zomato.csv", index=False)
print("✅ Cleaned dataset saved as cleaned_zomato.csv")
```

```python
# --- Clean 'rate_out_of_5' column ---
# Extract numeric part from strings like "4.1/5", "3.8 ", etc.
df['rate_out_of_5'] = df['rate_out_of_5'].astype(str).str.extract(r'(\d+\.?\d*)')

# Convert to numeric (floats), set invalids as NaN
df['rate_out_of_5'] = pd.to_numeric(df['rate_out_of_5'], errors='coerce')

# Keep only valid ratings between 0–5
df = df[df['rate_out_of_5'].between(0, 5)]

# Fill missing ratings with average rating
df['rate_out_of_5'] = df['rate_out_of_5'].fillna(df['rate_out_of_5'].mean())


# --- Clean 'avg_cost_two_people' column ---
# Remove symbols like ₹ and commas
df['avg_cost_two_people'] = df['avg_cost_two_people'].astype(str).str.replace("[^0-9]", "", regex=True)

# Convert to numeric
df['avg_cost_two_people'] = pd.to_numeric(df['avg_cost_two_people'], errors='coerce')

# Fill missing costs with median (robust for skewed values)
df['avg_cost_two_people'] = df['avg_cost_two_people'].fillna(df['avg_cost_two_people'].median()).astype(int)


# --- Clean 'num_of_ratings' column ---
df['num_of_ratings'] = pd.to_numeric(df['num_of_ratings'], errors='coerce').fillna(0).astype(int)


# ✅ Quick check
print(df[['restaurant_name', 'rate_out_of_5', 'num_of_ratings', 'avg_cost_two_people']].head(10))
print("\nNull values after cleaning:\n", df.isnull().sum())
```

```python
import ast

# --- Step 1: Check if reviews_list exists ---
if 'reviews_list' in df.columns:
    print("✅ reviews_list column found, parsing...")

    # Step 2: Convert stringified tuples into Python objects
    def parse_reviews(x):
        try:
            return ast.literal_eval(x)
        except (ValueError, SyntaxError):
            return []

    df['reviews_list'] = df['reviews_list'].astype(str).apply(parse_reviews)

    # Step 3: Explode so each review gets its own row
    reviews_df = df[['restaurant_name', 'reviews_list']].explode('reviews_list').dropna()

    # Step 4: Split tuple into (rating, review_text)
    reviews_df[['review_rating', 'review_text']] = pd.DataFrame(
        reviews_df['reviews_list'].tolist(), index=reviews_df.index
    )

    # Step 5: Clean review_rating
    reviews_df['review_rating'] = reviews_df['review_rating'].astype(str).str.extract(r'(\d+\.?\d*)')
    reviews_df['review_rating'] = pd.to_numeric(reviews_df['review_rating'], errors='coerce')

    # Step 6: Drop the old reviews_list column
    reviews_df = reviews_df.drop(columns=['reviews_list'])

    # Step 7: Clean text
    reviews_df['review_text'] = reviews_df['review_text'].astype(str).str.strip()

    # Save reviews to new file
    reviews_df.to_csv("/content/drive/MyDrive/Pan-India Zomato Restaurant Review Analytics/cleaned_reviews.csv", index=False)

    print("✅ Exploded reviews saved as cleaned_reviews.csv")
    print(reviews_df.head(10))

else:
    print("⚠️ reviews_list column not found in your dataset. Skipping...")
```

```python
# --- Clean 'online_order' column ---
df['online_order'] = df['online_order'].astype(str).str.strip().str.title()
df['online_order'] = df['online_order'].replace({'Yes':'Yes', 'No':'No'})

# --- Clean 'table_booking' column ---
df['table_booking'] = df['table_booking'].astype(str).str.strip().str.title()
df['table_booking'] = df['table_booking'].replace({'Yes':'Yes', 'No':'No'})

# --- Clean 'cuisines_type' (take only first cuisine if multiple) ---
df['cuisines_type'] = df['cuisines_type'].astype(str).str.split(',').str[0].str.strip().str.title()

# --- Clean 'area' ---
df['area'] = df['area'].astype(str).str.strip().str.title()

# --- Clean 'restaurant_name' ---
df['restaurant_name'] = df['restaurant_name'].astype(str).str.strip().str.title()

# ✅ Save final cleaned dataset
df.to_csv("/content/drive/MyDrive/Pan-India Zomato Restaurant Review Analytics/cleaned_zomato.csv", index=False)

print("✅ Final cleaned dataset saved as cleaned_zomato.csv")
print("Sample rows:\n", df.head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Top 10 most common cuisines
top_cuisines = df['cuisines_type'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="viridis")
plt.title("Top 10 Most Common Cuisines", fontsize=14)
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Type")
plt.show()


# 2. Online Order Availability
plt.figure(figsize=(6,6))
df['online_order'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("Set2"))
plt.title("Restaurants Offering Online Orders", fontsize=14)
plt.ylabel("")
plt.show()


# 3. Table Booking Availability
plt.figure(figsize=(6,6))
df['table_booking'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("Set3"))
plt.title("Restaurants Offering Table Booking", fontsize=14)
plt.ylabel("")
plt.show()


# 4. Distribution of Ratings
plt.figure(figsize=(8,6))
sns.histplot(df['rate_out_of_5'], bins=20, kde=True, color="blue")
plt.title("Distribution of Restaurant Ratings", fontsize=14)
plt.xlabel("Rating (out of 5)")
plt.ylabel("Number of Restaurants")
plt.show()


# 5. Average Cost for Two People by Area (Top 10 areas)
top_areas = df['area'].value_counts().head(10).index
avg_cost_area = df[df['area'].isin(top_areas)].groupby('area')['avg_cost_two_people'].mean().sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=avg_cost_area.values, y=avg_cost_area.index, palette="coolwarm")
plt.title("Average Dining Cost for Two (Top 10 Areas)", fontsize=14)
plt.xlabel("Average Cost (₹)")
plt.ylabel("Area")
plt.show()
```
